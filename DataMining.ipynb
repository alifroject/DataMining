```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([5,2,8,4])
y = np.array([1,6,7,8,])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot( x,  marker = 'o', ms = 10, mfc = 'r', mec = 'r', linestyle = "-", linewidth = 10) #marker size, marker face color,

plt.title("Ini judul", fontdict = font1, loc = "left")
plt.xlabel("Ini horizontal", fontdict = font2)
plt.ylabel("Ini vertikal", fontdict = font2)


plt.grid(axis = "y", color = 'r', linestyle = ':', linewidth = 2)


plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1,2,1) #It takes 1 row, 2 columns, and 1st sub
plt.plot(x,y, marker = "o", linestyle = "--", ms = 10, color = 'r')



x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])



plt.subplot(1,2,2) #It take 1 row, 2 columns, and second sub
plt.plot(x,y, marker = "o", linestyle = ":")

plt.show()

```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,2,1) #It takes 1 row, 2 columns, and 1st sub
plt.plot(x,y, marker = "o", linestyle = "--", ms = 10, color = 'r')



x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.title("Ini adalah sub plot", loc = "right")
plt.ylabel("Ada 2 subplot")

plt.subplot(2,2,4) #It take 1 row, 2 columns, and second sub
plt.plot(x,y, marker = "o", linestyle = ":")

plt.show()

```
